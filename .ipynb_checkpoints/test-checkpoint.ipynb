#### Working with Biopython

```python
#"data/gencode_5utrs_human.fa","fasta"
def readFASTA(filename):
    seqs=[]
    for seq_record in SeqIO.parse(filename,"fasta"):
        seqs.append(str(seq_record.seq))
    return seqs
humanseqs=readFASTA("data/gencode_5utrs_human.fa")
```

```python
seen = set()

for x in range(len(humanseqs)):
    if humanseqs[x] in seen:
        seen.add(humanseqs[x])
        humanseqs[x]=0
    else:
        seen.add(humanseqs[x])
```

```python
unique_humanseqs=[i for i in humanseqs if i != 0]
len(unique_humanseqs)
```

```python
np.array(unique_humanseqs)
group = np.array(['Human']*len(unique_humanseqs))
unique_humanseqs = np.vstack((unique_humanseqs,group)).T
```

```python
lengths=[]
for seq in unique_humanseqs:
    lengths.append(len(seq))
fig, axs = plt.subplots(nrows=1, ncols=1,layout='constrained' )
plt.hist(lengths, bins=300)
#plt.yscale('log')
#plt.xscale('log')
plt.title("Length Distribution")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.minorticks_on()
plt.show()
```
